# VCF files: file that stores genetic variants relative to a reference genome. 
Instead of storing all 3 billion bases, it only stores positions where the sample differs and the metadata about those differences (humans differ at ~4-5 million positions, tiny compared to 3 billion bases total). (that’s why the storage space so small). 

It has 2 main parts: Header (metadata) (these are the lines starting with ##) 
And then the data lines: each line = one variant. 
The columns represent chromosome, position (1-based), Variant ID, Reference base, Alternate base, Quality Score, Pass/fail Filter, extra annotations. 

gVCF: includes non-variant regions (VCF = only variant). It’s used for joint genotyping (explains increased storage space for gVCF). It includes them not by lsiting every base individually, but regions (ie 1-3: NON_REF [meaning same as reference, non-variant], then 4: T G [here is variant/change: point mutation]) . 

Variants in VCF (explained using examples): 
SNP: A G 
Insertion: REF = A, ALT = ATG 
Deletion: REF = ATG, ALT = A 



# BONUS: why do we use different file formats and when do we use which one? 
Modern sequencing generates multiple file types, each optimized for a different purpose: raw data storage, alignment, variant representation, and visualization. Their sizes and use cases differ dramatically. 
## 1.	FASTA:
 3GB/genome (uncompressed) and it contains reference DNA sequence (linear sequences of DNA). These files are used/required for alignment and variant calling 
## 2.	FASTQ: 
100 GB/genome, content is raw reads + quality scores. You use it for primary data from sequencing (sample) and it is needed for reprocessing with improved pipelines. 
## 3.	SAM:
 300 GB, for intermediate use only (temporary storage)
## 4.	BAM: 
100 GB, contains reads aligned to reference. It is used for variant calling (and visualization in genome browsers). (still smaller than FASTQ because of compression)
## 5.	CRAM: 
compressed alignment: 30GB, the content is the same as BAM but compressed relative to reference. Use it for long term storage of aligned reads. It is much smaller because it stores difference from the reference, not full sequences. 
## 6.	VCF 
(Variant Call Format): 100-200 MB /genome. Contains only variant positions (~4-5 million sites). Used for downstream analysis (annotation) and data sharing. Very small file size because only stores variants. 
## 7.	gVCF: 
1 GB/genome. Contains variants + non-variant regions. Used for joint genotyping across multiple samples. Role is intermediate for cohort analysis. Larger than vcf because it still includes information about all genomic regions. 

# NGS data analysis:
 shows/explains how the different file formats are used: fastq is where you save the “shredded” DNA fragments (raw reads) of the sequence/sample you want to look at (it also saves the quality scores). FASTA saves any genomes, relevant here: the reference genome (in pipelines). Then you do alignment (with tools like BWA) and match the reads to the reference producing a SAM -> converted to a BAM/CRAM file (Stores aligned reads). After alignment; do pre-processing (sort BAM, mark duplicates and recalibrate base quality) with tools like SAMtools, this step improves variant accuracy. Then you do variant calling (with GATK HaplotypeCaller) -> the reads are now aligned and now looks for differences /variants (uses statistical models to decide whether differences are real variants or sequencing errors): these differences are stored in a VCF file (contains only the variants). (after this you can do variant filtering, ie based on depth or quality and then annotate (add biological meaning), after which you can do interpretation/analysis and visualization). 

Storing called variants: VCF 
For archival purposes: CRAM or VCF are best. (SAM too large, BAM often replaced by CRAM for efficiency).
