Explain why the frequency table has a different shape from the sample table.
    - what one row represents in the sample table
    - what one row represents in the frequency table
    - why these two tables cannot be merged row by row

The frequency table has a different shape from the sample table because they describe the same data in 2 different levels of detail.  In the sample table each row is one biological sample vs in a frequency table, a row is one category (a label value) with its count (how often it occurs). So the frequency table is a compressed representation of the sample table. Essentially, the sample table is raw data (many rows, detailed) vs the frequency table is summarized data (fewer rows, aggregated). 
They are incompatible for row-wise merging because one row in the frequency table represents multiple rows in the sample table. The frequency table doesn't have individual identity information which the sample table has.
